### **📝 Task 14: Finding the maximum duration of a movie/tv shows available on Netflix.**

```python
df[['Minutes', 'Unit']] = df['Duration'].str.split(' ', expand=True)

df.head()
```

```python
df['Minutes'] = df['Minutes'].astype('int')

series = df['Category'] == 'TV Show'
movies = df['Category'] == 'Movie'

series_max = df[series][['Minutes', 'Unit']].max()
movies_max = df[movies][['Minutes', 'Unit']].max()

print(series_max)
print(movies_max)
```
